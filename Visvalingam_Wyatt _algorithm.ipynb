## Execution of Visvalingam-Whyatt Algorithm on GPS co-ordinates

```python
#Import necessary libraries

import pandas as pd
import numpy as np
from shapely.geometry import LineString
import matplotlib.pyplot as plt
from visvalingamwyatt import simplify
```

#### Applying Visvalingam-Whyatt Algorithm to random coordinates

```python
# Generate random coordinates
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.1, size=100)
coords_random = np.column_stack((x, y))
coords_random
```

```python

# Simplify the polyline
simplified_polyline = simplify(coords_random, 15)

# Create plot
fig, ax = plt.subplots()

# Plot original polyline
ax.plot(*zip(*coords_random), label='Original')

# Plot simplified polyline
ax.plot(*zip(*simplified_polyline), label='Simplified')

# Add legend and show plot
ax.legend()
plt.show()
```

#### Applying Visvalingam-Whyatt Algorithm to GPS coordinates

```python
# Load GPS track data from CSV file
data = pd.read_csv('gps_data.2.csv')
polyline = [(row['longitude'], row['latitude']) for _, row in data.iterrows()]
polyline
```

```python

# Simplify the polyline
simplified_polyline = simplify(polyline, 25)

# Create plot
fig, ax = plt.subplots()

# Plot original polyline
ax.plot(*zip(*polyline), label='Original')

# Plot simplified polyline
ax.plot(*zip(*simplified_polyline), label='Simplified')

# Add legend and show plot
ax.legend()
plt.show()
```
